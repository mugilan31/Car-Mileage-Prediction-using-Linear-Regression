# Train & test split

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = df.drop('mpg',axis=1)
y= df[['mpg']]
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=1) 
```

```python
X_train, X_test, y_train, y_test.shape
```

##  Linear Regression- Gradient Method

```python
regression_model = LinearRegression()

regression_model.fit(X_train,y_train)
```

```python
y_train_pred = regression_model.predict(X_train)
y_train_pred
```

```python
y_test_pred = regression_model.predict(X_test)
y_test_pred
```

```python
from sklearn.metrics import r2_score
r2_square_train = r2_score(y_train, y_train_pred)
print(r2_square_train)
```

```python
from sklearn.metrics import r2_score
r2_square_test = r2_score(y_test, y_test_pred)
print(r2_square_test)
```

## The Model is 83 - 84 % efficient Hence it is a Best Fit Model

```python
m = regression_model.coef_[0]
m
```

# Intercept

```python
c = regression_model.intercept_
c
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".
          format(col_name, regression_model.coef_[0][idx]))
```
